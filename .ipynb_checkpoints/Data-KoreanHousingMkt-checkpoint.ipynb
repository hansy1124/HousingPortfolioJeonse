```python
#1. Read housing price data
os.chdir(r'C:\Users\Owner\Dropbox\8. Python Space\Research1-HousingPortfolioJeonseData\Data\KoreanHousingMkt')
HPdf=pd.read_excel('KoreanHousingMkt.xlsx')
```

```python
A=list(HPdf.columns)
A[0]='Date' #Index itself is not mutable
HPdf.columns=pd.Index(A) 
HPdf.index=pd.PeriodIndex(pd.Index(HPdf['Date']), freq='M') #Period Index
```

```python
HPdf.columns
```

```python
#2. House Price & Rent Price & Jeonse Price Indices
```

```python
#2.1. Capital Area
HPdf[['Capital_Purchase','Capital_Jeonse','Capital_Rent']].plot()
plt.legend(loc=(1.04,0.5))
plt.title('Capital Area Price Dynamics')
plt.show()
```

```python
#2.2. Seoul Area
HPdf[['Seoul_Purchase','Seoul_Jeonse','Seoul_Rent']].plot()
plt.legend(loc=(1.04,0.5))
plt.title('Seoul Area Price Dynamics')
plt.show()
```

```python
#2.3. Country Area
HPdf[['Country_Purchase','Country_Jeonse','Country_Rent']].plot()
plt.legend(loc=(1.04,0.5))
plt.title('Country Area Price Dynamics')
plt.show()
```
