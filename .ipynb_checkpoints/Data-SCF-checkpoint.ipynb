# Data Structure of SCF

SCF2016 (p16i6.dta) is the main dataset that contains the all observations for all items.
- Each column means a question in questionaire.
- Each row means an observation for one respondent. \
: It seems there are five observations for each household, probably related to the imputation/need to study more

Codebook 2016 contains information about the questions and corresponding variables.

Following is the content from Codebook 2006.

## Important Variables
- YY1: Identifier for external users (Equal for all imputation for same person)
- Y1: Identifier for external users for imputation (YY1*10 + # imputation number) 

## Variable Names
- The codebook refers to the variables by the names they have in the version of the survey data set formatted for use with SAS. - These names consist of a number prefixed by an "X."  We have tried, insofar as it was possible, to retain the variable numbering system used in earlier SCFs.  
- Where the content of a variable has changed in a substantive way, we have assigned a new variable number.
- Each of the variables in the main data set has a "shadow" variable that describes--in most all cases--the original state of the variable (i.e., whether it was missing for some reason, a range response was given, etc.)
- The most important exception is reported values which have been imputed or otherwise altered to protect the privacy of respondents (see "DISCLOSURE REVIEW" below.)  
- Users who so desire may use the shadow variables to restore the data to something very close to their original condition. The shadow variables have the same numbers as the main variable, but have a prefix of "J."  
- A list of the values taken by the shadow variables is given in the section below entitled "DISCUSSION OF RANGE DATA COLLECTION AND J-CODES."

## Imputation
- To understand the imputed value (4 additional values other than the original value), read "Imputation" section in the codebook.
- For simple statistics such as average, just dividing the weight by 5 (the each imputation) should be okay
- Imputation is only for missing values.
- For more sophisticated analysis such as regression, refer to the codebook and resources in there

```python
#2.2. Main Dataset Construction
Mdf=SCF2016[b]
Mdf.columns=c
Mdf.columns
```

```python
#2.3. Make a 'RentOwnership' variable
#- 1, 3, 5 : 'ownership'
#- 2 : 'retner' 
#- others : 'other'

#2.4. Ownership Variable
C1=Mdf['HousingTenure']==5 #Townhouse
B1=Mdf['HousingTenure']==3 #Condo
A1=Mdf['HousingTenure']==1 #Own
A1=A1.astype(int)
B1=B1.astype(int)
C1=C1.astype(int)
D1=A1+B1+C1 #whether own or not?
D1.value_counts() #Worked well? Yes! - no 2 nor 3 & only 1 or 0
```

```python
#2.4. Make a 'Renter' variable
E1=Mdf['HousingTenure']==2
E1=E1.astype(int)
E1.value_counts() #worked well? no 2 or 3.. only 1 or 0
```

```python
#2.5. Make a 'Housing Status' variable - 1 means ownership, 2 means rent, 0 means others
Mdf['RentOwnership']=D1+E1*2
Mdf['RentOwnership'].value_counts() #worked well
```

```python
#2.6. Age group variable
Mdf['Age'].describe()
```

```python
#2.6. Continued
bins = [18, 30, 40, 50, 60, 70, 80, 120]
#labels = ['18-29', '30-39', '40-49', '50-59', '60-69', '70-79','80+']
labels1 = [1, 2, 3, 4, 5, 6, 7]
Mdf['AgeGroup']=pd.cut(Mdf.Age, bins, labels = labels1, include_lowest = True)
Mdf['AgeGroup'].value_counts()
```

```python
#3.4. Plot the Ratio Graphs

fig, ax = plt.subplots(figsize=(10,7))  

months = df['HousingTenure'].drop_duplicates()
margin_bottom = np.zeros(len(df['AgeGroup'].drop_duplicates()))
colors = ["#006D2C", "#31A354","#74C476"]

for num, month in enumerate(months):
    values = list(df[df['HousingTenure'] == month].loc[:, 'Ratio'])

    df[df['HousingTenure'] == month].plot.bar(x='AgeGroup',y='Ratio', ax=ax, stacked=True, 
                                    bottom = margin_bottom, color=colors[num], label=month)
    margin_bottom += values
plt.title("Housing Tenure By Age in 2017")
plt.legend(loc=(1.04,0.5))
plt.xticks(rotation=360)
plt.xlabel('Age Group')
plt.show()
```
