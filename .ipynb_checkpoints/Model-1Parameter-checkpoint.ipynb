## 1. Life Parameter

```python
#1.1. Longevity
LT=10   # Live LT years. From 20 to 19+LT / Starting at age 20

#1.2. Survival Probability Conditional on Survival at t 
lambdaPS=list() 
lambdaPS.append(1) # Probability of surviving at 20, which is 1
for i in range(0,LT-1):
    lambdaPS.append(lambdaPS[i]*0.999)

#At the end of the age, household must die
lambdaPS[LT-1]=0
lambdaPS=pd.DataFrame(lambdaPS)    
    
#1.3. Actual Survival Probability
FPS=list()
FPS.append(1)
for i in range(0,LT-1):
    FPS.append(lambdaPS.iloc[i]*FPS[i])    
    
#1.4. Age Grid 
AgeGrid=list()
AgeGrid.append(20)
for i in range(0,LT-1):
    AgeGrid.append(i+21)
    
#1.5. Probability Graph
fig = plt.figure(figsize=(10,5))

ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

ax1.plot(AgeGrid, FPS)
ax1.title.set_text('Probability of Surviving - F')

ax2.plot(AgeGrid, lambdaPS)
ax2.title.set_text('Conditional Probability of Surviving - Lambda')

plt.show()
```

## 2. Household Problem Parameters

- For each line, the value next to '#' is the benchmark number.
- Actual value can be different from it depending on the type of exercise that I want to do

```python
#2.1. Expected Asset Return
R_f=1.03 #1.03 risk free rate #deposit rate - https://ko.tradingeconomics.com/south-korea/deposit-interest-rate
mu_r=0.035 #0.015 #risk premium - http://news.khan.co.kr/kh_news/khan_art_view.html?art_id=201808301028001
mu_h=0.02 #0.025 #expected growth of house price - http://www.index.go.kr/potal/main/EachDtlPageDetail.do?idx_cd=1240

#2.2. Household Problem Parameter [Housing]
#2.2.1. Jeonse
Jbar=0.6638 #0.6638 #ratio of Jeonse deposit to house price -> Refer to the R.file (https://www.r-one.co.kr/rone/resis/common/main/main.do)
delta_J=0.5 #0.3 #downpayment ratio for Jeonse - https://m.post.naver.com/viewer/postView.nhn?volumeNo=17720223&memberNo=25828090
phi_J=0.015 #0.004 #Jeonse contract cost - https://m.post.naver.com/viewer/postView.nhn?volumeNo=16281311&memberNo=7322715
DTI_J=0.4 #0 DTI Constraint for Jeonse - no constraint in the market actually
#2.2.2. Rent
tau_r=Jbar*0.055 #Jbar*0.07 #전월세 전환율 - http://r-one.co.kr/rone/resis/statistics/statisticsViewer.do?menuId=HOUSE_21121&autoExec=Y
#2.2.3. Ownership&Purchase
chi=0.003 #0.003 maintenance fee - Need more info 
delta=0.6 #0.5 #downpayment ratio for home purchase - https://m.post.naver.com/viewer/postView.nhn?volumeNo=17720223&memberNo=25828090
phi_b=0.015 #0.005 #House purchase cost - https://m.post.naver.com/viewer/postView.nhn?volumeNo=16281311&memberNo=7322715
phi=0.01  #0.005 #Selling cost - https://m.post.naver.com/viewer/postView.nhn?volumeNo=16281311&memberNo=7322715
Hbar=500 #Minimum house value (lower than it, rent or jeonse)
DTI_H=0.5 #DTI Constraint for Purchase - https://m.post.naver.com/viewer/postView.nhn?volumeNo=17720223&memberNo=25828090

#2.3. Household Problem Parameter [Household Utility Function&Bequest Motives]
#2.3.1. Utility Function
omega=0.33 #Housing versus non-durable consumption
rho=1/3 #Inverse of the elasticity of intertemporal substitution 
gamma=4 #CRRA risk aversion parameter
beta=0.97 #Discount factor
xi=0.01 #Probability of moving - Vestman 2019 different for each age
lambda_r=0.65 #ratio of retirement income to the last working age income
#2.3.2. Bequest
alpha_f=0.35 #Bequest strength - (annuity factor)
beta_b=0.91 #Bequest strength - (discount factor)
T_b=3 #Bequest strength - (time period)
b_b=(beta*(1-beta_b**T_b)/(1-beta_b))*(((1-omega)**(1-omega))*(omega**omega)/(tau_r**omega)) #constant for bequest term

#2.4. Labor Income Process After the retirement, change in labor income growth rate - Arbitrarily set now
#2.4.1. Retirement Income Parameter
Lage=8  # Retirement Age = 20 + Lage
#2.4.2. Labor Income Process After the Retirement
#- With PSID data, we need to estimate the cubic polynomial of labor income process 
#- but I here assume the simple concave increasing fuction
LIncomeGrowthProcess=list()
LIncomeProcess=list()
LIncomeProcess.append(100) #Labor income profile at age 20 - 100.

for i in range(AgeGrid[0],Lage+AgeGrid[0]): # From 20 to 20+Lage-1
    LIncomeGrowthProcess.append(0.02) #net groth rate 2%
for i in range(Lage+AgeGrid[0],LT+AgeGrid[0]): # From 20+Lage to 20+LT-1
    LIncomeGrowthProcess.append(0)

#Once you calibrate labor income growth rate, put that vector into LIncomeGrowth Process, 
#-> Then it will generate the labor income process by folloiwng code.    
k=1
for i in range(AgeGrid[0],Lage+AgeGrid[0]):
    LIncomeProcess.append(LIncomeProcess[-1]*(1+LIncomeGrowthProcess[k]))
    k=k+1
LIncomeProcess.append(LIncomeProcess[-1]*lambda_r)

for i in range(Lage+AgeGrid[0]+1,LT+AgeGrid[0]-1):
    LIncomeProcess.append(LIncomeProcess[-1]*(1+LIncomeGrowthProcess[k]))
LIncomeGrowthProcess=pd.DataFrame(LIncomeGrowthProcess)

plt.plot(AgeGrid,LIncomeGrowthProcess)
plt.title('Labor Income Growth Process')
plt.show()
print(len(LIncomeGrowthProcess))

plt.plot(AgeGrid,LIncomeProcess)
plt.title('Labor Income Process')
plt.show()
print(len(LIncomeProcess))
```

## 3. Parameter for Shock Process

```python
#Based on Vestman 2019

#3.1. Idiosyncratic shock for each shock 
delta_v=0.1283 #standard deviation of idiosyncratic labor income shock
delta_ep=0.2 #standard deviation of idiosyncratic stock market return
delta_n=0.20 #standard deviation of housing return

#3.2. Correlation between shocks
corrHS=0.1 #correlation between housing return and stock return 
eta_en=(delta_n/delta_ep)*(1/(1-corrHS)) #coefficient in front of stock return shock so that it goes into housing return equation for correlation.
corrYS=0.10 #correlation between labor income and stock return
corrHY=0.14 #correlation between housing return and labor income

#3.3. Coefficient parameters for acheiving above correlation values
#-> Complicated function. First solve eta_nn by fsolve and solve eta_eo

#Paramter list
list1=[corrHY, corrYS, delta_n, delta_v, delta_ep, eta_en]

#Function solve (Online appendix 2 of Vestman 2019 later <- We have much easier way actually)
def etanfunction (eta_nn, params=list1):
    corrHY=list1[0]
    corrYS=list1[1]
    delta_n=list1[2]
    delta_v=list1[3]
    delta_ep=list1[4]
    eta_en=list1[5]
    return (corrHY**2)*((delta_n**2)*(delta_v**2)+(eta_nn**2)*(delta_n**4)+(eta_en**2)*(delta_ep**2)*(delta_v**2)+(eta_en**2)*(eta_nn**2)*(delta_ep**2)*(delta_n**2))-(delta_n**4)*(eta_nn**2)-((1-corrHY**2)*(delta_ep**4)*(eta_nn**2)-(corrHY**2)*(delta_ep**2)*(delta_n**2))*((corrYS**2)/(1-corrYS**2))*((delta_v**2+(eta_nn**2)*(delta_n**2))/(delta_ep**2))-(2*(delta_n**2)*(delta_ep**2)*(eta_nn)*(eta_en))*((corrYS)/(math.sqrt(1-corrYS**2)))*((math.sqrt(delta_v**2+(eta_nn**2)*(delta_n**2)))/(delta_ep))

#Coefficient for shock correlated to housing return and exist in labor income shock part
eta_nn=scipy.optimize.fsolve(etanfunction, corrHY)
eta_nn=float(eta_nn)

#Coefficient for shock correlated to stock return and exist in labor income shock part
eta_eo=((corrYS)/math.sqrt(1-corrYS**2))*(math.sqrt(delta_v**2+(eta_nn**2)*delta_n**2)/(delta_ep)) 

# For actual shock process generation, go to Shocks.m.ipynb
```
